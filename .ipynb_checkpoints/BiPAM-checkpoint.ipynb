```python
import os
import sklearn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
theta = np.linspace(np.pi/180*50,0,1000) # rad
alpha = 10*np.pi/180 # rad

d = 1 
k = np.cos(alpha)

L0 = 6*10**-3 # m
L = 2*L0*np.sin(theta)+d*10**-3 # mm
L2 = 2*L0*np.sin(theta) # mm
D = 54.5*10**-3 # m
Di = L0*np.cos(theta)/3 # m
Ds = D-Di/3

n = 14 # ea
```

```python
L_tot = n*L
L_tot = (L_tot-L_tot[-1])*1000
L_tot2 = n*L2
L_tot2 = (L_tot2-L_tot2[-1])*1000

#L_tot2 - L_tot
L
```

```python
P = -40*10**3 # Pa

A_c = np.pi*D**2/4

A = L0**2*np.sin(2*theta)/2
L_arc = np.pi*np.sqrt(Ds**2+(L/np.pi)**2)

dDs = L0*np.sin(theta)/3
dL = 2*L0*np.cos(theta)
dA = L0**2*np.cos(2*theta)
dL_arc = (np.pi**2*Ds*dDs+L*dL)/L_arc

dV = A_c*dL - dA*L_arc - A*dL_arc

F = -P*dV/dL

plt.plot(L_tot,F)
```

```python
F=[0]*7
for i in range(0,6):
    
    P = -1 * 10* (i)*10**3 # Pa

    A_c = np.pi*D**2/4

    A = L0**2*np.sin(2*theta)/2
    L_arc = np.pi*np.sqrt(Ds**2+(L/np.pi)**2)

    dDs = L0*np.sin(theta)/3
    dL = 2*L0*np.cos(theta)
    dA = L0**2*np.cos(2*theta)
    dL_arc = (np.pi**2*Ds*dDs+L*dL)/L_arc

    A = A_c*dL 
    B = dA*L_arc
    C = A*dL_arc
    dV = A_c*dL - dA*L_arc - A*dL_arc

    F[i] = -P*dV/dL

    plt.plot(L_tot,F[i])
    F[6] = dV/dL
    plt.grid(True)
```

```python
data = {
    'length': L_tot,    
    '0 kPa' : F[0],
    '10 kPa' : F[1],
    '20 kPa' : F[2],
    '30 kPa' : F[3],
    '40 kPa' : F[4],
    '50 kPa' : F[5],
    'dV/dL' : F[6]

}
frame = pd.DataFrame(data)
```

```python
frame
```

```python
frame.to_csv("ActuatorModelDisplacement.csv")
```

```python
A2 = A_c*dL/dL*10000
B2 = dA*L_arc/dL*10000
C2 = A*dL_arc/dL*10000
D2 = B2+C2 
E2 = A2-D2
F2 = A2*0.7

plt.plot(L_tot,A2,'k-', label='Cylinder')
# plt.plot(L_tot,-D2,'g-',label='Helical shape')
plt.plot(L_tot,E2,'r-',label='BiPAM Contraction')
plt.plot(L_tot,F2,'b-',label='Pneumatic cylinder with friction')

plt.xlabel('Displacement (mm)')
plt.ylabel('Force (N)')

plt.xlim([0, 130])
plt.ylim([-30, 30])

plt.legend()
plt.grid(True)
plt.show()
```
